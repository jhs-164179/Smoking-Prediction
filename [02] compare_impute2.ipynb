```python
import warnings

warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


plt.rcParams['font.family'] ='NanumGothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
def plot_dist_multi(dfs, na_series, labels, title="Multiple Distributions", figsize=(12, 6)):
    """
    dfs: 여러 개의 DataFrame 리스트
    na_series: 결측값이 있는 컬럼 목록
    labels: 각 DataFrame의 레이블 리스트
    title: 그래프 제목
    figsize: 그래프 크기 조정
    """
    num_features = len(na_series.index)
    row = 2
    # col = (num_features // row) + (num_features % row > 0)
    col = 3

    fig, ax = plt.subplots(nrows=row, ncols=col, figsize=figsize)
    ax = ax.flatten()
    colors = sns.color_palette("Set1", len(dfs))

    for idx, feature in enumerate(na_series.index):
        for df, label, color in zip(dfs, labels, colors):
            # sns.histplot(df[feature], ax=ax[idx], label=label, 
            #              color=color, alpha=0.5)
            sns.kdeplot(df[feature], ax=ax[idx], color=color, lw=2.5, label=label, common_norm=False)

        ax[idx].set_title(feature)
        ax[idx].legend()

    fig.suptitle(title, fontsize=14)
    plt.tight_layout()
    plt.show()
```

```python
ori = pd.read_csv('./survey.csv')
imp0 = pd.read_csv('./impute_set/imp0.csv')
imp1 = pd.read_csv('./impute_set/imp1.csv')
imp2 = pd.read_csv('./impute_set/imp2.csv')
imp3 = pd.read_csv('./impute_set/imp3.csv')
imp4 = pd.read_csv('./impute_set/imp4.csv')
imp5 = pd.read_csv('./impute_set/imp5.csv')
imp6 = pd.read_csv('./impute_set/imp6.csv')
```

```python
dfs = [ori, imp0, imp1, imp2]
labels = ['origin', 'fillna0', 'mode median', 'statics per cage']
plot_dist_multi(dfs, na_values, labels, figsize=(20, 6), title='Compare Distributions per Imputation Methods')
```
