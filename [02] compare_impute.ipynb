```python
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

warnings.filterwarnings('ignore')
plt.rcParams['font.family'] ='NanumGothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
def plot_dist(df, na_series, title='Origin Distribution'):
    cnt = 0
    row = 2
    col = (len(na_series.index)//row)
    fig, ax = plt.subplots(nrows=row, ncols=col, figsize=(10, 6))
    for i in range(row):
        for j in range(col):
            if cnt == len(na_series):
                break
            sns.histplot(df[na_series.index[cnt]], ax=ax[i,j], kde=True)
            ax[i,j].set_title(na_series.index[cnt])
            cnt += 1
    fig.suptitle(title)
    plt.tight_layout()
    plt.show()
```

```python
ori = pd.read_csv('./survey.csv')
imp0 = pd.read_csv('./impute_set/imp0.csv')
imp1 = pd.read_csv('./impute_set/imp1.csv')
imp2 = pd.read_csv('./impute_set/imp2.csv')
imp3 = pd.read_csv('./impute_set/imp3.csv')
imp4 = pd.read_csv('./impute_set/imp4.csv')
imp5 = pd.read_csv('./impute_set/imp5.csv')
imp6 = pd.read_csv('./impute_set/imp6.csv')
```

```python
na_values = ori.isna().sum()
na_values = na_values[na_values!=0].sort_values(ascending=False)
na_values = na_values[:6]
```

```python
plot_dist(ori, na_values)
```

```python
plot_dist(imp0, na_values, title='Fillna(0)')
```

```python
plot_dist(imp1, na_values, title='Fillna --> median and mode')
```

```python
plot_dist(imp2, na_values, title='Fillna --> statics per cage')
```

```python
plot_dist(imp3, na_values, title='MICE')
```

```python
plot_dist(imp4, na_values, title='Iterative Imputer')
```

```python
plot_dist(imp5, na_values, title='KNN Imputer')
```

```python
plot_dist(imp6, na_values, title='Missing Forest')
```
